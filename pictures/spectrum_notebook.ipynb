```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
large_scales = lambda x: x**(-1)
inertial_range= lambda x: x**(-5/3)
dissipation_range = lambda x: x**(-1.6) + 30**(-5/3)-30**(-1.6)

x1 = np.arange(0.4,1.1,0.1)
y1 = [large_scales(x) for x in x1]

x2 =np.arange(1,30.1,0.1)
y2 = [inertial_range(x) for x in x2]

x3 =np.arange(30,70,1)
y3 = [dissipation_range(x) for x in x3]
```

```python
fig=plt.figure(figsize = (8,6))
ax = plt.subplot2grid((1,1),(0,0))

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim([0.3,200])
ax.set_ylim([0.0001,10])
ax.yaxis.set_ticks([])
ax.yaxis.set_ticklabels([])
ax.xaxis.set_ticks([0.5,12])
ax.xaxis.set_ticklabels(['$1/L$', '$1/\overline{\Delta}$'],fontsize=15)


ax.set_title('Turbulent kinetic energy spectrum')
ax.set_xlabel("$\log(k)$",fontsize=15)
ax.set_ylabel("$\log(e_{k})$",fontsize=15)

ax.plot(x1,y1,c='k')
ax.plot(x2,y2,c='k')
ax.plot(x3,y3,c='k')

ax.annotate('$k^{-5/3}$',xy=(8,0.08), xytext=(8,0.1),fontsize=15)
ax.annotate('inertial \n range',xy=(1,0.001), xytext=(3,0.0006),fontsize=15)
ax.annotate(s='', xy=(1,0.0004), xytext=(30,0.0004), arrowprops=dict(arrowstyle='<->'))
ax.annotate('dissipation',xy=(70,0.001), xytext=(70,0.005),fontsize=15, arrowprops=dict(facecolor='black', shrink=0.05),
            va="center", ha="center")

plt.vlines(12,0.0001,12**(-5/3), linestyles='dashed')
plt.vlines(30,0.0001,30**(-5/3), linestyles='dashed', lw=1)
plt.vlines(1,0.0001,1, linestyles='dashed', lw=1)
plt.tight_layout()
plt.savefig("ek_spectrum.pdf")
```
